```python
#Dependences
import csv
import os 
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read data file
artist_data = pd.read_csv("artists.csv")
artist_data.head(10)
```

```python
#Convert artist data into a pandas dataframe and remove unnecessary columns
artist_data_df= pd.DataFrame(artist_data,columns=['artist_lastfm','listeners_lastfm','scrobbles_lastfm'])

#drop missing values
artist_data_df.dropna


data_df = pd.DataFrame()

# dropping duplicates and finding unique values for artist name
for column in artist_data_df.columns[0:]:
    data = artist_data_df[column].drop_duplicates()
    df1 = pd.DataFrame(data)
    data_df = data_df.append(df1)

#add header to index
data_df.index.name = 'id'

#show data
data_df.head()
```

```python
#slice dataset from over 1million rows to 20000
top20000_df = data_df[0:20000]
top20000_df.shape
top20000_df.head()
```

```python
top20000_df.to_csv('top20000.csv')
```

```python
#show list of top 20000
top20000_df.artist_lastfm.unique()
```

```python
billboard_data = pd.read_csv("Billboard.csv")
billboard_data.head(101)
```

```python
billboard_data_df= pd.DataFrame(billboard_data)
billboard_data_df.dropna(how='all')
billboard_data_df.head()
```

```python
#Converter

rds_connection_string = "postgres:postgres@localhost:5432/music_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#create table for top20000
top20000_df.to_sql(name='top20000', con=engine, if_exists='append', index=True)
```

```python
#create table for Billboard
billboard_data_df.to_sql(name='Billboard', con=engine, if_exists='append', index=True)
```
